## What can I observe in this dataset?

1. Remove first column and reset index
2. Sales_ID: There is a redundant space and random exclamation mark in each entry

    Is it possible the Sales_Branch information shifted to Sales_ID?

### Step1: Drop the unwanted column

```python
df2 = df.drop(columns=['Unnamed: 0'])
df2
```

### Step2: Slice and remove " " and "!" in Sales_ID

```python
# This can be done using Pandas rstrip()
df2['Sales_ID'] = df2['Sales_ID'].str.rstrip('!')
df2
```

```python
# Use str.strip() to remove any unnecessary characters just in case there is any
df2['Sales_ID'] = df2['Sales_ID'].str.strip()
df2
```

### Step3: Fill in missing data in column Sales_Branch

```python
# Double check Sales_Branch column are all null
df2.Sales_Branch.isnull()
```

The first 2 letters as shown in Sales_ID belongs to Sales_Branch, there are 2 things need to be done
    1. First thing is to slice 
    2. Then move to Sales_Branch column

For details can read throught the API:
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.slice.html

**Parameters**

start:
Start position for slice operation.

stop:
Stop position for slice operation.

step:
Step size for slice operation

```python
# The first 2 letters as shown in Sales_ID belongs to Sales_Branch
# First thing is to slice 
# Then move to Sales_Branch column
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.slice.html

df2['Sales_Branch'] = df2['Sales_ID'].str.slice(stop=2)
df2
```

### Step4: Reformatting Recent_Sales_Date column

The current Recent_Sales_Date column shows both the date and time (including time zone) of the transaction. I would prefer changing it to "year-month'day" format as this will be useful for future time-series analysis.

This can be achieved in 2 steps:
    1. As mentioned above, we could use the str.slice() method to slice the wanted section
    2. Next step we should do a data type check to make sure the column is in datetime format

```python
# Step 1 slicing
df2['Recent_Sales_Date'] = df2['Recent_Sales_Date'].str.slice(stop=10)
df2
```

```python
# data type check
df2.info()
# Recent_Sales_Date shows data type is and object
# Change dtype to datetime
```

```python
df2['Recent_Sales_Date'] = pd.to_datetime(df2['Recent_Sales_Date'])
df2.info()
```

### Step5: Reformatting Product_Description column

Let's revisit the content in Product_Descirption

```python
df2['Product_Description'][0]
```

We observed a long description in the Product_Description column which consists of two parts:

- The company description
- List of products

Here I will split the sentence and separate them into 2 columns - Company_Description and Products. This can be done using pd.Series.str.split( ) syntax

```python
df2['Product_Description'].str.split(' Product: ', expand=True)
```

```python
# Assign new column names and concat to the orgiinal DataFrame
df2[['Company_Description', 'Product']] = df2['Product_Description'].str.split(': ', expand=True)
df2
```

```python
# add comma in between product names, that would help with product counts if needed
df2['Product'] = df2['Product'].str.split()
df2
```

### Step 6: Include Product_count column

```python
# add a new column for Product_count

df2['Product_count'] = df2['Product'].str.len()
df2
```

```python
df2.sort_values(by=['Product_count'], ascending = False)
```

You can always re-organize the DataFrame columns based on your preference.
